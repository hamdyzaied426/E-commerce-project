# Data Wrangling

```python
# to find missing values
df.isnull().sum()
```

```python
# to show data null plot 
sns.heatmap(df.isna())
```

```python
# drop the Column besacuse it is empty
df = df.drop(['Column1'],axis=1)
```

```python
# drop the column that hasnot data
df = df.dropna()
```

```python
# to make sure there is no null values
df.isnull().sum()
```

```python
# to find duplicated
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].dtype
```

```python
df = df.drop(['Year','Month'],axis=1)
```

```python
df['Year'] = (df['Date']).dt.year
df['Month'] = (df['Date']).dt.month
df['Day'] = (df['Date']).dt.day
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,8))
df.plot()
```

```python
# to 
def find_Age_Customer(age):
    if age<18:
        return 1
    elif age>= 18 and age<40:
        return 2
    elif age>= 40 and age<60:
        return 3
    else:
        return 4

df['Age_Customer'] = df['Customer Age'].apply(lambda x: find_Age_Customer(x)) 
df.head()
```

```python
df['Age_Customer'] = df['Age_Customer'].map({1 : 'boys' , 2 : 'Youth', 3 : 'Adults' , 4 : 'Seniors'})
```

```python
df['Net profit'] = df['Revenue'] - df['Cost']
df.head()
```

```python
df.info()
```

```python
# merge the object columns
df_object = df.select_dtypes(include=['object'])
df_object
```

```python
# describe the object columns
df_object.describe()
```
